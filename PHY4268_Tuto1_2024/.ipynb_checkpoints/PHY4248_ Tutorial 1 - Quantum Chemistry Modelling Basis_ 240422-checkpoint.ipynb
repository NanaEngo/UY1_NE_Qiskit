## Structure lectronique des molcules en premire et seconde quantifications

### Hamiltonien d'une molcule en premire quantification

L'Hamiltonien d'une molcule constitue de noyaux $K$ et d'lectrons $N$ est
\begin{equation*}
\mathtt{H}=\mathtt{T}_e+\mathtt{T}_n+\mathtt{U}_{en}+\mathtt{U}_{en}+\mathtt{U}_{nn},
\end{equation*}
o
- $\mathtt{T}_e= -\sum_i\frac{\hbar^2}{2m_e}\nabla^2_i$ est l'nergie cintique lectronique;
- $\mathtt{T}_n=-\sum_I\frac{\hbar^2}{2M_I}\nabla^2_I$ est l'nergie cintique nuclaire;
- $\mathtt{U}_{en}= -\sum_{i,I}\frac{e^2}{4\pi\epsilon_0}\frac{Z_I}{|\mathbf{r}_i-\mathbf{R}_I|}$ est la rpulsion Coulombienne entre les lectrons et les noyaux;
- $\mathtt{U}_{ee}=+\frac{1}{2}\sum_{i\neq j}\frac{e^2}{4\pi\epsilon_0}\frac{1}{|\mathbf{r}_i-\mathbf{r}_j|}$ est la rpulsion Coulombienne entre les lectrons eux-mmes;
- $\mathtt{U}_{nn}=+\frac{1}{2}\sum_{I\neq J}\frac{e^2}{4\pi\epsilon_0}\frac{Z_IZ_J}{|\mathbf{R}_I-\mathbf{R}_J|}$ est la rpulsion Coulombienne entre les noyaux eux-mmes.

$M_I$, $\mathbf{R}_I$ et $Z_I$ ($I=1,2$) dsignent la masse, la position et le numro atomique du *I*-me noyau, et $\mathbf{r}_i$ est la position du *i*-me lectron. Par souci de concision, nous travaillons en units atomiques, o l'unit de longueur est $a_0 = 1$ Bohr ($0.529167~\mathtt{T}imes~10^{-10}$ m), l'unit de masse est la masse lectronique $m_e$ et l'unit d'nergie est 1 Hartree ($1~\textrm{Hartree}~= ~{e^2}/{4\pi\epsilon_0 a_0}~=~27.2113~\textrm{eV}$). Dnotant $M_I'=M_I/m_e$, les diffrents termes de l'Hamiltonien molculaire en units atomiques s'crit
\begin{equation*}
\begin{split}
\mathtt{T}_e&=-\sum_i\frac{\nabla^2_i}{2}, \qquad\mathtt{T}_n= -\sum_I\frac{\nabla^2_I}{2M_I'}, 
 \qquad\mathtt{U}_{en}= - \sum_{i,I}\frac{Z_I}{|\mathbf{r}_i-\mathbf{R}_I|},\\
\mathtt{U}_{ee}&=+\frac12\sum_{i\neq j}\frac{1}{|\mathbf{r}_i-\mathbf{r}_j|}
=+\sum_{i,j>i}\frac{1}{|\mathbf{r}_i-\mathbf{r}_j|},\\
\mathtt{U}_{nn}&=+\frac{1}{2}\sum_{I\neq J}\frac{Z_IZ_J}{|\mathbf{R}_I-\mathbf{R}_J|}
=+\sum_{I,J>I}\frac{Z_IZ_J}{|\mathbf{R}_I-\mathbf{R}_J|}.
\end{split}
\end{equation*}

### Approximation de Born-Oppenheimer

Nous nous intressons principalement  la structure *lectronique* de la molcule. Comme $M'\sim10^3$, nous appliquons l'**approximation de Born-Oppenheimer**, en traitant les noyaux comme des charges ponctuelles classiques. En consquence, le terme d'nergie cintique nuclaire $T_n$ est nglig, $\mathtt{U}_{nn}(\mathbf{R})$ est constant et l'Hamiltonien molculaire est approch par un Hamiltonien lectronique paramtr par $\mathbf{R}$,
\begin{equation*}
\begin{split}
\mathtt{H}(\mathbf{R}) &= \mathtt{U}_{nn}+\underset{\text{One-Electron Operators}}{\mathtt{T}_e + \mathtt{U}_{en}(\mathbf{R})} + \underset{\text{Two-Electron Operator}}{\mathtt{U}{ee}}\\
&= \mathtt{U}_{nn}+\underset{\text{One-Electron Operator}}{\sum_i h(i)} + \underset{\text{Two-Electron Operator}}{\sum_{i\ne j}v(i, j)}.
\end{split}
\end{equation*}

### nergie potentiel de surface

En vertu de l'quation de $\mathtt{H}(\mathbf{R})$ prcedente, l'quation de Schrdinger indpendante du temps non relativiste crit
\begin{equation*}
\mathtt{H}(\mathbf{R})\psi_i(\mathbf{R},\mathbf{r})=E_i(\mathbf{R})\psi_i(\mathbf{R},\mathbf{r}),
\end{equation*}
o $\psi_i(\mathbf{R},\mathbf{r})$ dsigne l'un des tats propres et $E_i(\mathbf{R})$ correspond aux surfaces d'nergie potentielle molculaire (PES), ce qui est important pour comprendre les processus chimiques tels que la formation et la rupture de liaison. En rgle gnrale, la prcision du calcul du PES (Potential Energy Surface) doit tre limite  la *prcision chimique* $\epsilon=1 \textrm{kcal}\textrm{mol}^{-1}(\sim 1.59\times 10^{-3}\textrm{Hartrees},\, 43.3\textrm{meV})$.

L'nergie de l'tat fondamental est donne par
\begin{equation}\tag{1} 
E_0 = \frac{\langle\psi_0|\mathtt{H}|\psi_0\rangle}{\langle\psi_0|\psi_0\rangle},
\end{equation}
o $|\psi_0\rangle$ est l'tat fondamental du systme. Cependant, la dimensionnalit de ce problme crot exponentiellement avec le nombre de degrs de libert. Pour rsoudre ce problme, nous aimerions prparer $|\psi_0\rangle$ sur un calculateur quantique et mesurer directement la valeur moyenne Hamiltonienne (ou $E_0$). Alors, comment pouvons-nous faire cela concrtement?

Un bon point de dpart pour rsoudre ce problme est la mthode Hartree-Fock (HF), autrement nous allons prendre comme fonction d'tat de rffrence la fonction d'tat HF. Cette mthode rapproche un problme  N corps en N problmes  un corps o chaque lectron volue dans le champ moyen des autres.  Il est  noter que le dnominateur de l'Equation (1) n'est ncessaire que quand la fonction d'etat HF n'est pas normalise au dpart. Une facon d'amliorer cette fonction d'tat HF est d'utiliser une combinaison linaire des fonction d'tat. Ce qui sera fait  travers les bases chimiques.


### Dterminant Slater

Un dterminant de Slater est un produit anti-symtrique d'un ou de plusieurs spin orbitals. Pour un systme  $N$-electron, le dterminant de Slater s'crit
\begin{equation*}
\begin{split}
\psi(\mathbf{r}_0, \mathbf{r}_1, \ldots, \mathbf{r}_{N-1})\equiv |\phi_{M-1},\cdots,  \phi_1, \phi_0 \rangle 
&=   \frac{1}{\sqrt{N!}}
\begin{vmatrix} \phi_0(\mathbf{r}_0) & \phi_1(\mathbf{r}_0) & \cdots & \phi_{M-1}(\mathbf{r}_0) \\
      \phi_0(\mathbf{r}_1) & \phi_1(\mathbf{r}_1) & \cdots & \phi_{M-1}(\mathbf{r}_1) \\
      \vdots & \vdots & \ddots & \vdots \\
      \phi_0(\mathbf{r}_{N-1}) & \phi_1(\mathbf{r}_{N-1}) & \cdots & \phi_{M-1}(\mathbf{r}_{N-1})
\end{vmatrix} \\
&\equiv |M-1,\dots, 1, 0 \rangle,
\end{split}
\end{equation*}
 L'change des positions de deux lectrons est quivalent  l'change de deux ranges du dterminant de Slater, ce qui modifie le signe de la fonction d'tat. Cela fournit la symtrie d'change correcte pour la fonction d'tat fermionique. Bien que le nombre de spin-orbitales considres, $M$, soit gnralement plus lev que le nombre d'lectrons dans le systme, $N$, les lectrons ne peuvent occuper que $N$ des spin-orbitales dans un dterminant Slater donn. En consquence, le dterminant de Slater ne contient que les spin-orbitales occupes $N$.

### Exemple d'illustration

Pour rendre les correspondances plus claires, nous examinons comment une fonction d'tat sera reprsente pour un systme quantique fictif. On considre les orbitales de spin $|A_\uparrow\rangle,\,|A_\downarrow\rangle,\,|B_\uparrow\rangle,\,|B_\downarrow\rangle$. Nous sommes libres de dfinir arbitrairement l'tat Hartree-Fock de notre systme fictif et de choisir les deux lectrons dans l'orbitale $|A\rangle$. Nous sommes intresss par la fonction d'tat lorsque la composante $z$ du spin est nulle.

Labellisons chaque orbitale ainsi qu'il suit:

\begin{align*}
 &|A_\uparrow\rangle=|00\rangle=|\mathbf{0}\rangle,  &&|A_\downarrow\rangle=|01\rangle=|\mathbf{1}\rangle,
&&|B_\uparrow\rangle=|10\rangle=|\mathbf{2}\rangle,  &|B_\downarrow\rangle=|11\rangle=|\mathbf{3}\rangle.
\end{align*}

L'tat de Hartree-Fock possde les deux lectrons dans les orbitales $|A\rangle$. Un tat HF incorrectement symtris serait donc $|A_\uparrow\rangle_1|A_\downarrow\rangle_2 = |\mathbf{0}\rangle_1|\mathbf{1}\rangle_2$, o les indices indiquent l'lectron dcrit par chaque orbitale. La fonction d'tat HF correctement antisymtrise serait 

\begin{equation*}
\begin{split}
    |\Psi_{\mathrm{HF}}\rangle =  \frac{1}{\sqrt{2}}
		\begin{vmatrix}
		A_\uparrow(\mathbf{x}_0) & A_\downarrow(\mathbf{x}_0)\\
		A_\uparrow(\mathbf{x}_1) & A_\downarrow(\mathbf{x}_1)\\
		\end{vmatrix}
		&= \frac{1}{\sqrt{2}}  \Big( A_\uparrow(\mathbf{x_0}) A_\downarrow(\mathbf{x_1}) - A_\downarrow(\mathbf{x_0}) A_\uparrow(\mathbf{x_1})  \Big)\\
&=\frac{1}{\sqrt{2}} (|\mathbf{0}\rangle_1|\mathbf{1}\rangle_2 - |\mathbf{1}\rangle_1|\mathbf{0}\rangle_2 ).
\end{split}
\end{equation*}

Si nous considrons maintenant les excitations au-dessus de l'tat HF, alors une fonction d'tat gnrale avec $s_z=0$ qui a t correctement antisymtrise est
donne par

\begin{equation*}
\begin{aligned}
|\Psi\rangle =& \frac{\alpha}{\sqrt{2}}(|\mathbf{0}\rangle_1|\mathbf{1}\rangle_2 - |\mathbf{1}\rangle_1|\mathbf{0}\rangle_2 ) 
+\frac{\beta}{\sqrt{2}} (|\mathbf{2}\rangle_1|\mathbf{3}\rangle_2 - |\mathbf{3}\rangle_1|\mathbf{2}\rangle_2 ) \\
+& \frac{\gamma}{\sqrt{2}} (|\mathbf{0}\rangle_1|\mathbf{3}\rangle_2 - |\mathbf{3}\rangle_1|\mathbf{0}\rangle_2 ) + \frac{\delta}{\sqrt{2}} (|\mathbf{1}\rangle_1|\mathbf{2}\rangle_2 - |\mathbf{2}\rangle_1|\mathbf{1}\rangle_2 ).
\end{aligned}
\end{equation*}

Comme nous avons $N=2$ lectrons et $M=4$ spin-orbitales, nous pouvons voir que nous avons seulement besoin de $N \lceil\mathrm{log_2}(M)\rceil = 2 \times \lceil\mathrm{log_2}(4)\rceil = 4$ qubits pour stocker la fonction d'tat.


## Seconde quantification - Construction d'un oprateur Hamiltonien fermionique

### Gnration des intgrales molculaires

L'Hamiltonien  est exprim dans la base des solutions de la mthode HF, galement appeles Orbitales Molculaires (OM) :
$$
\mathtt{H}_{elec}=\sum_{pq} h_{pq} a^{\dagger}_p a_q + 
\frac{1}{2} \sum_{pqrs} h_{pqrs}  a^{\dagger}_p a^{\dagger}_q a_r  a_s
$$
avec 
* les **intgrales  1 lectron**
$$
h_{pq} = \int \phi^*_p(r) \left( -\frac{1}{2} \nabla^2 - \sum_{I} \frac{Z_I}{|\mathbf{r}_i-\mathbf{R}_I|} \right)   \phi_q(r)d\mathbf{r}
$$
qui dcrivent lnergie cintique des lectrons individuels et leurs interactions Coulombiennes avec les champs lectriques du noyau ;
* et **intgrales  2 lectrons**
$$
h_{pqrs} = \int \frac{\phi^*_p(r_1)  \phi^*_q(r_2) \phi_r(r_2)  \phi_s(r_1)}{|\mathbf{r}_1-\mathbf{r}_2|}d\mathbf{r}_1d\mathbf{r}_2,
$$
dcrivent les interactions Coulombiennes entre les lectrons;
* $a^\dagger_p$ l'**oprateur de cration** d'un lectron sur le spin-orbite $p$;
* $a_q$ l'**oprateur d'annihilation** d'un lectron sur sur le spin-orbite $q$;
* $a^{\dagger}_p a_q $ est l'**oprateur d'excitation**, qui excite un lectron du spin-orbite occup $\phi_q$ vers le spin-orbite inoccup $\phi_p$.

Les MO ($\phi_u$) peuvent tre occups ou virtuels (inoccups). Un MO peut contenir 2 lectrons. Ces MO sont les solutions de la mthode HF, o chaque lelectron est soumis au champ moyen cr par les autres lectrons. Cependant, dans ce qui suit, nous travaillons en fait avec des orbitales de spin qui sont associes  un spin up ($\alpha$) d'lectron spin down ($\beta$). Ces deux spins sont galement communment dsigns par $\alpha$ et $\beta$, respectivement. Ainsi, les orbitales de spin peuvent contenir un lectron ou tre inoccupes.

Dans le tableau ci-dessous resume les correspondances premire-seconde quantification des divers lments du Hamiltonien molculaire. 
$$
\begin{array}{|l|l|l|}\hline
\textbf{Approximation BO}  &  \textbf{Combinaison en 2e quantification} &  \textbf{Description}
\\\hline
-\frac12\sum_i\nabla^2_i & -\frac12\sum_{p,q}\langle p|\nabla^2_p|q\rangle a^\dagger_p a_q & \mathtt{T}_e \\\hline
 -\sum_{i,I}\frac{Z_I}{|\mathbf{r}_i-\mathbf{R}_I|} &
 -\sum_{p,q} \langle p|\frac{Z_I}{|\mathbf{r}_p-\mathbf{R}_I|}|q\rangle a^\dagger_p a_q
 & \mathtt{U}_{en}\\\hline
\sum_{i,j>i}\frac{1}{|\mathbf{r}_i-\mathbf{r}_j|} & 
\sum_{p,q,r,s}\langle pq \big|\frac{1}{|\mathbf{r}_i-\mathbf{r}_j|}\big|rs\rangle a^\dagger_p a^\dagger_q a_r a_s & 
 \mathtt{U}_{ee}\\\hline
\end{array}
$$

*Un des avantages apport par le formalisme de seconde quantification est que la proprit d'anti-symtrie des fonctions d'tat pour l'change de fermions identiques, prise en compte manuellement dans une approche de premire quantification en considrant les dterminants de Slater, est automatiquement impose par le relations d'anti-commutation des oprateurs de cration et d'annihilation. Cependant, le prix  payer est de travailler avec un nombre non fixe de particules, qui reste de toute faon une quantit conserve.*

### Reprsentation nombre d'occupations

En seconde quantification, le dterminant de Slater est reprsent par un **vecteur nombre d'occupations**
\begin{equation*}
|f\rangle =|f_{M-1},\dots,f_{i-1},f_i,f_{i+1},\dots, f_0\rangle, 
\end{equation*}
avec
- $f_i=1$ lorsque le spin-orbital $\phi_i$ est occup et donc prsent dans le dterminant de Slater;
- $f_i = 0$ lorsque le $\phi_i$ spin-orbital est vide et donc absent du dterminant de Slater.

- Le vecteur $|f_i\rangle$ est appel **vecteur nombre d'occupations de l'orbite fermionique i**, parce que
\begin{align*}
& |0\rangle=\begin{pmatrix}1\\0\end{pmatrix}:\text{tat vide}, & |1\rangle=\begin{pmatrix}0\\1\end{pmatrix}:\text{tat occup}.
\end{align*}

Dans la deuxime reprsentation de quantification, au lieu de poser la question *Quel lectron est dans quel tat?*, nous posons la question *combien de particules y a-t-il dans chaque tat?*.

### Oprateurs fermioniques

Les oprateurs fermioniques $a^\dagger_i$ et $a_i$ obissent aux relations d'anticommutation fermioniques suivantes:
\begin{align*}
&\{a_p,a^\dagger_q\} = a_pa^\dagger_q + a^\dagger_q a_p = \delta_{pq}, 
&\{a_p, a_q \} = \{a^\dagger_p, a^\dagger_q\} = 0 .
\end{align*}
La nature fermionique des lectrons implique que les fonctions d'tat  plusieurs lectrons doivent tre antisymtriques par rapport  l'change de particules. Cela se reflte dans la manire dont les oprateurs de cration fermionique et d'annhilation agissent sur les dterminants $|f\rangle$:
\begin{equation*}
\begin{aligned}
&a_i|f_{M-1},\dots,f_{i-1},f_i,f_{i+1},\dots, f_0\rangle = \delta_{f_i,1}
(-1)^{\sum_{m<i}f_m}|f_{M-1},\dots,f_{i-1},f_i\oplus 1,f_{i+1},\dots, f_0\rangle,\\
&a^\dagger_i|f_{M-1},\dots,f_{i-1},f_i,f_{i+1},\dots, f_0\rangle = \delta_{f_i,0}
(-1)^{\sum_{m<i}f_m}|f_{M-1},\dots,f_{i-1},f_i\oplus 1,f_{i+1},\dots, f_0\rangle`,  \\
&a|0\rangle=a^\dagger|1\rangle=0.
\end{aligned}
\end{equation*}
- Le terme de phase $p_i=(-1)^{\sum_{m<i}f_m}$ dsigne la parit qu'impose l'anti-symtrie d'change des fermions.
    - $p_i=-1$ si le nombre d'lectrons est impair dans ces orbitales de spin,
    - $p_i= 1$ si le nombre d'lectrons est pair dans ces orbitales de spin.

- Le symbole $\oplus$ reprsente l'addition modulo 2, i.e.,  $1 \oplus 1=0 $ et $0 \oplus 1=1$.

- L'oprateur de cration $a^\dagger_i$ met un lectron dans une orbitale inoccupe *i* ou dans $|f_i\rangle$.

- L'oprateur d'annihilation $a_i$ supprime un lectron dans une orbitale occupe *i* ou en $|f_i\rangle$.

- $a_i^2 =(a^\dagger_i)^2= 0$ pour tout $i$. On ne peut pas crer ou anantir un fermion dans le mme mode deux fois.

Par example,
\begin{align*} 
&a^\dagger_1 |0\rangle_1 = |1\rangle_1,&&a^\dagger_1 |1\rangle_1 =(a^\dagger_i)^2|0\rangle_1= 0,&& a_1 |1\rangle_1 =|0\rangle_1,&a_1 |0\rangle_1 =a_1^2|1\rangle_1=0. 
\end{align*}
Notez qu'ici $a^\dagger_1|1\rangle_1=0$ et $a_1 |0\rangle_1=0$ signifie le vecteur zro et non $|0\rangle_1$. 

En utilisant galement de tels oprateurs, nous pouvons exprimer
\begin{equation*}
|0\rangle |1\rangle |1\rangle |0\rangle = a^\dagger_1a^\dagger_2 |0\rangle^{\otimes 4}.
\end{equation*}

 L'oprateur d'occupation orbitale est donn par
\begin{align*}
&n_i = a^\dagger_i a_i, &n_i |f_{M-1},\dots,f_i,\dots, f_0\rangle= f_i |f_{M-1},\dots,f_i,\dots, f_0\rangle,
\end{align*}
et il compte le nombre d'lectrons dans une orbitale donne, c'est--dire
\begin{align*} 
&n_i |0\rangle_i = 0, &n_i |1\rangle_i = |1\rangle_i. 
\end{align*}

## Calcul du champ moyen ou mean-field

Le calcul du champ moyen est gnralement la premire tape de toutes les approches post Hartree-Fock. Les approches en champ moyen rsolvent l'quation de Schrdinger en faisant la moyenne du terme de rpulsion lectron-lectron de manire auto-cohrente ou SCF. En dautres termes, chaque lectron ninteragit pas explicitement avec les autres lectrons.

Les trois mthodes HF communment utilises sont (voir la figure ci-dessous) :

![HF_Orb.png](./Graphics/HF_Orb.png)

1. **RHF (Restricted Hartree  Fock)** utiliee pour des molcules  couches pleines ou fermes. Les spin-orbitales sont soit $\alpha$, soit $\beta$ et tous les orbitales sont doublement occups par des spin-orbitales $\alpha$ et $\beta$;</br> 
2. **ROHF (Restricted Open-Shell HartreeFock)** utilise pour des molcules  couches ouvertes o le nombre d'lectrons les orbitales n'est pas le mme. ROHF utilise autant que possible les orbitales doublement occupes et les orbitales une fois occupes par les lectrons non-aparis;
3. **UHF (Unrestricted Hartree-Fock)** utilise pour des molcules  couches ouvertes o le nombre d'lectrons les orbitales n'est pas le mme. Les orbitales UHF peuvent avoir avoir des spin $\alpha$ ou $\beta$, mais les orbitales $\alpha$ et $\beta$ peuvent avoir des composants spatiales diffrents.

Les quivalents DFT sont,

4. **RKS (Kohn-Sham restreint)**,
6. **ROKS (Restricted Open-Shell Kohn  Sham)**,
5. **UKS (Kohn-Sham sans restriction)**.

**Python-based Simulations of Chemistry Framework (PySCF)** est un programme de chimie computationnelle *ab initio* implment nativement dans le langage de programme Python. Le package vise  fournir une plate-forme simple, lgre et efficace pour le dveloppement et le calcul de codes de chimie quantique. Il fournit diverses fonctions pour faire, au niveau non relativiste, 
* la thorie de Hartree-Fock,
* la Second-order MllerPlesset perturbation theory (MP2),
* la thorie fonctionnelle de la densit (DFT),
* la Multi-configuration self-consistent field (MCSCF),
* la thorie des clusters coupls (CC).

 PySCF est utilis dans la plupart des packages de calculs quantiques pour les calculs HF (premire quantification) et diverses tapes des transformations de seconde quantification.

Il est  noter que la mthode SCF ou Self Consistent Field est une mthode variationnelle o la fonction d'essai est un dterminant de Slater ou une combinaison linaire de ces dterminants.

Il est  noter que les termes HOMO et LUMO sont utiliss en chimie quantique pour dsigner les orbitales molculaires particulires d'une molcule. 

* **HOMO (Highest Occupied Molecular Orbital)** est la plus haute orbitale molculaire occupe. Il reprsente l'orbitale o se trouvent les lectrons les plus nergtiques et est donc la plus proche de la couche de valence. Les lectrons prsents dans l'HOMO sont gnralement impliqus dans les ractions chimiques, les liaisons avec d'autres atomes ou molcules et les interactions lectroniques.

* **LUMO (Lowest Unoccupied Molecular Orbital)** est la plus basse orbitale molculaire non occupe. Elle reprsente l'orbitale qui peut accepter des lectrons lors de ractions chimiques ou d'interactions avec d'autres molcules. Le LUMO est gnralement d'une nergie plus leve que l'HOMO et est impliqu dans les processus de transfert d'lectrons, de formation de liaisons chimiques ou de ractions de dissociation.

* La diffrence d'nergie entre l'HOMO et le LUMO, appele le **gap HOMO-LUMO**, est galement un facteur important dans la dtermination des proprits d'absorption et d'mission de lumire des molcules organiques, ce qui est important en chimie des matriaux et en spectroscopie.

### Cration d'un objet mean-field (mf) avec `scf.RHF()`

Pour la plupart des usages, les types de calculs SCF pertinents sont:

1. RHF (Restricted Hartree  Fock)
2. UHF (Hartree-Fock sans restriction)
3. ROHF (Restricted Open-Shell HartreeFock)


L'initialisation de l'objet champ moyen pour ces six types de calculs est illustre ci-dessous. Ces classes ncessitent au moins un argument,  savoir un objet molcule, `mol`.

1. `mf=scf.RHF(mol)`
2. `mf=scf.UHF(mol)`
3. `mf=scf.ROHF(mol)` (o mol est dfinie comme tant  couche ouverte (open-shell), mol.spin!=0)
4. `mf=scf.RKS(mol)`
5. `mf=scf.UKS(mol)`
6. `mf=scf.ROKS(mol)` (o mol est dfinie comme open-shell, mol.spin!=0)

On peut aussi spcifier les paramtres de calculs du mf.

```python
mf=scf.RHF(DWat_mol)

mf.conv_tol=1e-12 # the difference in the SCF energy (in Hartrees) between two successive cycles
mf.conv_tol_grad=1e-8 # the root-mean-square of the orbital gradient
mf.direct_scf_tol=1e-13 #the threshold for discarding integrals
mf.init_guess='atom' #  vital for the efficient completion of the SCF procedure
mf.max_cycle=100 # the maximum number of SCF cycles that should be performed before the calculation terminates. For systems that are notoriously difficult to converge, the value should be increased to 100 or even 1000
mf.max_memory=4000 #determines the maximum amount of memory (in Megabytes) that PySCF is allowed to utilize during the SCF procedure
mf.verbose=5 #controls the print level for the mean-field object (in the 'Dwat_HF.txt' defined above)

mf.kernel()
```

La mthode SCF la plus largement utilise en chimie quantique est sans aucun doute la thorie de la fonctionnelle de la densit (DFT), et excuter un calcul DFT dans PySCF est presque aussi simple que d'excuter un calcul Hartree-Fock.

Afin d'accder  des fonctions supplmentaires qui ne relvent pas de la norme SCF/HF, il est ncessaire d'importer le module dft (thorie fonctionnelle de la densit).

```python
from pyscf import dft

mfd=dft.KS(DWat_mol)

mfd.conv_tol=1e-12
mfd.conv_tol_grad=1e-8
mfd.direct_scf_tol=1e-13
mfd.init_guess='1e'
mfd.max_cycle=100
mfd.max_memory=8000

mfd.xc='wB97X-D' # la fonctionnelle

mfd.kernel()
```

## Choix du set de bases

L'quation de Schrdinger  plusieurs lectrons est gnralement rsolue sur une base d'orbitales molculaires exprimes sous la forme d'une **combinaison linaire de ses orbitales atomiques (LCAO, Linear combination of atomic orbitals)**, 
$$
     \phi_a(\mathbf{r}) =
     \sum_A^\mathrm{atomes}
     \sum_\alpha C^A_{\alpha a}
     \chi^A_{\alpha}(\mathbf{r} - \mathbf{R}_A)
$$
Les coefficients LCAO, utiliss pour construire des orbitales molculaires (MO), sont obtenus grce  la minimisation de l'nergie totale, selon le principe variationnel. Pour plus de commodit, la base est gnralement reprsente par des fonctions centres sur les atomes (les positions atomiques) pour les systmes molculaires, ou par des fonctions d'tat plane en physique du solide.

Les deux classes d'orbitales de base approximatives couramment utilises sont :
* les **Slater-type orbitals (STOs)** bases sur le dterminant de Slater, construitent avec des fonctions exponentielles dcroissantes $e^{-\alpha r}$,
* et les **orbitales cartsiennes de type Gaussien (GTO)**, construitent avec des fonctions Gaussiennes $e^{-\alpha r^2}$. Lutilisation de ces fonctions Gaussiennes est bien plus efficace que la stratgie STO pour calculer les nombreuses intgrales de chevauchement ncessaires  la chimie quantique.

Ces deux types de fonctions de base peut tre combin comme **STO-nG (Slater-type orbital-n Gaussians)**, o n est le nombre de Gaussiennes utilises pour faire les approximations.

Les ensembles de base sont construits  partir de nombreuses primitives Gaussiennes. La notation de Pople est l'une des notations les plus explicites pour les ensembles de bases:
- le trait d'union (-) spare la rfrence aux orbitales centrales et de valence;
- les chiffres reprsentent le nombre de primitives Gaussiennes dans une fonction ;
- il y a autant de chiffres qu'il y a de fonctions pour chaque orbitale atomique (zeta $\zeta$) ;
- la lettre G fait rfrence  l'utilisation de fonctions Gaussiennes.

 titre d'exemple, l'ensemble de base double zta `3-21G` utilise trois primitives Gaussiennes pour chaque orbitale centrale et un ensemble de deux plus une primitives Gaussiennes pour les orbitales de valence. Plus d'informations, d'exemples et de discussions sur les GTO peuvent tre trouvs dans ce [lien](https://chem.libretexts.org/Courses/Pacific_Union_College/Quantum_Chemistry/11%3A_Computational_Quantum_Chemistry/11.02%3A_Gaussian_Basis_Sets).

Gnralement, on adopte des ensembles identiques de fonctions de base pour tous les atomes d'un lment donn et les ensembles adopts pour les diffrents lments doivent avoir une prcision quilibre. Une dclaration telle que  *nous avons utilis la base 6-31G dfinie dans le calcul*  signifie qu'un ensemble quilibr d'ensembles d'AO a t adopt pour tous les lments.

Ci-dessous, des exemples densembles de base populaires utiliss en chimie quantique.


```python
basis_sets = [
    "STO-3G",           # Simple zeta, minimal basis
    "3-21G",            # Double zeta with 3 Gaussian primitives
    "6-31G",            # Double zeta with 6 Gaussian primitives
    "6-31G*",           # Double zeta with 6 Gaussian primitives
    "6-31G(d,p)",       # Polarization functions (+ 5 d-orbitals for all atoms except H, +3 p-orbitals for H atoms) added
    "6-311G(d,p)",      # Triple zeta with polarization functions
    "6-311+G(d,p)",     # Triple zeta with polarization functions and diffuse functions
    "cc-pvqz",          # Quadruple zeta
    "cc-pv5z",          # Quintuple zeta
    "def2-SVP",         # Double zeta with polarization functions
    "def2-SVPD",         # Double zeta with polarization functions and diffuse functions
    "def2-TZVP",        # Triple zeta with polarization functions
    "def2-TZVPD",       # Triple zeta with polarization functions and diffuse functions
    "def2-TZVPP",       # Triple zeta with polarization functions and diffuse functions
    "def2-TZVPPD",      # Triple zeta with polarization functions, diffuse functions, and d functions
    "def2-QZVP",        # Quadruple zeta with polarization functions
    "def2-QZVPP",       # Quadruple zeta with polarization functions and diffuse functions
    "def2-QZVPPD",      # Quadruple zeta with polarization functions, diffuse functions, and d functions
    "def2-QZVPD",       # Quadruple zeta with polarization functions and diffuse functions
    "def2-QZVPPD",      # Quadruple zeta with polarization functions, diffuse functions, and d functions
]
```

Because the basis set is an approximation, it is highly desirable to be able to control its accuracy in order to make tradeoffs between the cost of the calculation and the accuracy of the obtained results. Accordingly, modern basis sets typically come in families of varying size: the smallest sets enable quick but qualitative calculations, while the larger sets enable quantitative computations at the cost of more computer time. In contrast to traditional basis sets, modern basis set families allow for a cost-efficient approach to the complete basis set limit, at which point the error in the one-electron basis set no longer affects the calculation. Note that also other types of basis sets than Gaussians may be used for quantum chemistry.

The Karlsruhe def2 family of Gaussian basis sets,227 which are a good all-round choice for general chemistry as they are available for the whole periodic table up to radon (Z = 86). The Karlsruhe def2 sets come in three levels of accuracy. Split-valence (SV) basis sets are the smallest reasonable basis set for general applications. The def2-SVP basis is a SV basis set with polarization (P) functions, and is similar in size to the 6-31G** also known as the 6-31G(d,p) basis set. Like 6-31G**, the def2-SVP set can also be used without polarization functions on hydrogen atoms; this basis is called def2-SV(P), it is smaller than the 6-31G* basis, and it is often useful for quick qualitative/semi-quantitative calculations. For more quantitative calculations, the def2 series also contains a triple- valence polarization set (def2-TZVP) as well as a quadruple- valence polarization set (def2-QZVP), which typically suffice for achieving the complete basis set limit in HF and DFT calculations. Calculations
at post-HF levels of theory, however, require larger basis sets with additional polarization functions; the def2-TZVPP and def2-QZVPP basis sets exist for this purpose. Diffuse functions (D) are necessary for the proper description of anions as well as to model e.g. electric polarizabilities; sets are likewise available at all levels of accuracy (def2-SVPD, def2-TZVPD, def2-TZVPPD, def2-QZVPD, def2-QZVPPD) for this purpose.

La fonction d'tat pour une molcule est une approximation lors de l'utilisation d'un ensemble de base finie,
$$
\Phi_{\text{approx}} = \sum_{i=0}^N c_i |i \rangle
$$

En vertu du principe variationnel, en considrant que $\Phi_{\text{approx}}$ est normalise, l'nergie approxime est suprieure ou gale  la vritable nergie de l'tat fondamentale,
$$
\langle \Phi_{\text{approx}} | \hat{H} | \Phi_{\text{approx}} \rangle = \epsilon_{\text{approx}} \geq  \epsilon_{\text{true}}
$$

A mesure que l'ensemble de base s'agrandit, l'approximation rsultante  se rapproche de la vritable fonction d'tat,
$$
\Phi_{\text{approx}} = \sum_{i=0}^{\infty} c_i |i \rangle \approx \Psi_{\text{true}}
$$

Cependant, augmenter la taille de l'ensemble de base a pour effet d'augmenter le nombre de paramtres  optimiser, tendant ainsi les ressources de calculs requises dans l'espace (espace de stockage) et en temps (nombre d'oprations). Il est souvent judicieux d'quilibrer la taille de l'ensemble de base pour optimiser le temps de rsolution par rapport  la prcision cible ($|\epsilon_{\text{approx}} -  \epsilon_{\text{true}}|$), puisque l'augmentation de la complexit de l'ensemble de base produit des rendements dcroissants en termes de gain de prcision.

Une liste complte des ensembles de bases disponibles pris en charge dans [PySCF](https://pyscf.org/) est disponible sur ce [lien](https://pyscf.org/_modules/pyscf/gto/basis.html). De plus en plus de bases sont disponibles pour la communaut, et le choix d'un ensemble de bases peut s'avrer une tche difficile. La thorie stipule qu'une base infinie est le moyen d'obtenir de vrais calculs des fonctions d'tat. En pratique, il n'est pas possible d'effectuer les calculs avec des ensembles de bases infinis et d'obtenir des vraies fonction d'tat. On peut cependant slectionner des ensembles de bases finis et adapter nos approches aux cas d'utilisation. En rgle gnrale, 
* un double zta (par exemple `6-31G` ou `cc-pvdz`) est la taille de base minimale requise pour obtenir des nergies semi-quantitatives;
* pour les molcules prsentant une diffrence d'lectrongativit leve entre les atomes, l'ajout de fonctions de polarisation (par exemple `6-31G**` ou `6-31G(d,p)`) est recommand;
* les fonctions diffuses (par exemple `6-31G+(d,p)`) doivent tre incluses lorsqu'il s'agit d'anions. Les fonctions diffuses sont des fonctions avec de petits coefficients (petit $\alpha$ dans $e^{-\alpha r^2}$) qui peuvent s'adapter  la densit lectronique loin du noyau;
* les ensembles de bases des potentiels de noyau effectif (Effective Core Potential, ECP) sont ncessaires lorsqu'il s'agit d'atomes plus lourds avec des effets relativistes et des effets non triviaux (par exemple, l'atome d'iode est souvent modlis dans l'ensemble de base `LANDL2DZ`).

### Effet de l'ensemble de base sur les calculs d'nergie totale et le temps de calculs

```python
import time

mf_energies = list()
mf_times = list()
nb_prim = list()

# Perform a Mean-Field calculation for each basis set
for bs in basis_sets:

    # Measure execution time
    start = time.time()
    DWat_mol.basis = bs
    DWat_mol.build()

    mf = scf.RHF(DWat_mol)
    mf.kernel()
    end = time.time()

    nb_prim.append(DWat_mol.npgto_nr())
    mf_energies.append(mf.e_tot)
    mf_times.append(end-start)
```

```python
# Create the results dataframe
import pandas as pd

df_HF = pd.DataFrame({"Basis":basis_sets, 
                      'Nb of GTO primitives':nb_prim, 
                      'Total energy':mf_energies,
                      "Time":mf_times})

df_HF
```
